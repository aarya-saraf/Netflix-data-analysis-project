creating a function which will convert numerical data to categorical data 

```python
def categorized_col(df,col,labels):
    edges=[df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]
    df[col]=pd.cut(df[col],edges,labels=labels,duplicates='drop')
    return df
```

convertion of vote average from numerical to categorical so that it can be analyzed easily

```python
labels=['not_popular','below_average','average','popular']
categorized_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```

checking if their is any null value present or not

```python
df.dropna(inplace=True)
df.isna().sum()
```

removing wide spaces in genre and expanding each genre to its own row

```python
import ast

# If Genre values look like strings (e.g. "[Action, Adventure]")
df['Genre'] = df['Genre'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else x)

# If Genre values are comma-separated strings
df['Genre'] = df['Genre'].astype(str).str.split(',')

# Remove spaces
df['Genre'] = df['Genre'].apply(lambda x: [i.strip() for i in x] if isinstance(x, list) else x)

# Expand each genre to its own row
df = df.explode('Genre').reset_index(drop=True)

df.head()
```

checking the categories from the genre column

```python
df['Genre']=df['Genre'].astype('category')
df['Genre'].dtypes
```
